```python
import numpy as np

def compute_tf(sentences):
    vocabulary = sorted(set(word for sentence in sentences for word in sentence.lower().split()))
    word_to_index = {word: index for index, word in enumerate(vocabulary)}
    tf_matrix = np.zeros((len(sentences), len(vocabulary)), dtype=float)
    for i, sentence in enumerate(sentences):
        words = sentence.lower().split()
        for word in words:
            if word in word_to_index:
                tf_matrix[i][word_to_index[word]] += 1/len(words)
    return tf_matrix, vocabulary
```

```python
def compute_idf(sentences, vocabulary):
    num_sentences = len(sentences)
    idf_vector = np.zeros(len(vocabulary), dtype=float)
    word_to_index = {word: index for index, word in enumerate(vocabulary)}
    for word in vocabulary:
        count = sum(1 for sentence in sentences if word in sentence.lower().split())
        idf_vector[word_to_index[word]] = np.log(num_sentences / (1 + count))+1
    return idf_vector
```

```python
def compute_tf_idf(sentences):
    tf_matrix, vocabulary = compute_tf(sentences)
    idf_vector = compute_idf(sentences, vocabulary)
    tf_idf_matrix = tf_matrix * idf_vector
    return tf_idf_matrix, vocabulary

sentences = [
    "The cat sat on the mat",
    "The dog sat on the log",
    "The cat and the dog are friends"
]
tf_idf_matrix, vocabulary = compute_tf_idf(sentences)
print("Vocabulary:", vocabulary)
print("TF-IDF Matrix:\n", tf_idf_matrix)
```
